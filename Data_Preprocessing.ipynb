```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('atp_matches_qual_chall_2023.csv')
df1.head()
```

```python
df2 = pd.read_csv('atp_rankings_current.csv')
df2.head()
```

```python
df3 = pd.read_csv('atp_matches_2023.csv')
df3.head()
```

```python
df4 = pd.read_csv('atp_matches_futures_2023.csv')
df4.head()
```

```python
# Count the number of matches in each filtered DataFrame
num_matches_df1 = len(df1)
num_matches_df3 = len(df3)
num_matches_df4 = len(df4)

# Create a bar chart
labels = ['Challenge and Qualification', 'Tournament Level', 'Future']
num_matches = [num_matches_df1, num_matches_df3, num_matches_df4]

plt.figure(figsize=(7, 6))
plt.bar(labels, num_matches, color=['blue', 'green', 'orange'])
plt.xlabel('Match Types')
plt.ylabel('Number of Matches')
plt.title('Distribution of match data divided by tournament level')
plt.show()
```
